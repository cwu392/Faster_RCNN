```python
img_dir = '/data/a2301133/TT100K/data/'
with open("tt100k_train.txt", "wb") as f: 
    for x in data['imgs']:
        imgs = data['imgs'][str(x)]
        img_path = data['imgs'][str(x)]['path']
        path = img_dir + img_path
        if 'train' in img_path:
            bbox = data['imgs'][str(x)]['objects']
            for y in bbox:
                sub_bbox = y['bbox']
                xmax, xmin, ymax, ymin = sub_bbox['xmax'], sub_bbox['xmin'], sub_bbox['ymax'], sub_bbox['ymin']
                xmax, xmin, ymax, ymin = int(xmax), int(xmin), int(ymax), int(ymin)
                category = y['category']
                line = path + ',' + str(xmin) + ',' + str(ymin) + ',' + str(xmax) + ',' + str(ymax) + ',' + category + "\n"
                f.write(line)
f.close()
```

```python
img_dir = '/data/a2301133/TT100K/data/'
with open("tt100k_all.txt", "wb") as f: 
    for x in data['imgs']:
        imgs = data['imgs'][str(x)]
        img_path = data['imgs'][str(x)]['path']
        path = img_dir + img_path
        bbox = data['imgs'][str(x)]['objects']
        for y in bbox:
            sub_bbox = y['bbox']
            xmax, xmin, ymax, ymin = sub_bbox['xmax'], sub_bbox['xmin'], sub_bbox['ymax'], sub_bbox['ymin']
            xmax, xmin, ymax, ymin = min(int(xmax), 2048), max(int(xmin), 0), min(int(ymax), 2048), max(int(ymin), 0)
            category = y['category']
            line = path + ',' + str(xmin) + ',' + str(ymin) + ',' + str(xmax) + ',' + str(ymax) + ',' + category + "\n"
            f.write(line)
f.close()
```

```python
with open("tt100k_test.txt", "wb") as f: 
    for x in data['imgs']:
        imgs = data['imgs'][str(x)]
        img_path = data['imgs'][str(x)]['path']
        path = img_dir + img_path
        if 'test' in img_path:
            bbox = data['imgs'][str(x)]['objects']
            for y in bbox:
                sub_bbox = y['bbox']
                xmax, xmin, ymax, ymin = sub_bbox['xmax'], sub_bbox['xmin'], sub_bbox['ymax'], sub_bbox['ymin']
                category = y['category']
                line = path + ',' + str(xmin) + ',' + str(ymin) + ',' + str(xmax) + ',' + str(ymax) + ',' + category + "\n"
                f.write(line)
f.close()
```
